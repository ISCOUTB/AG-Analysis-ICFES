```python
# Sometimes it just reads it in under than a minute, but at the cost of a lot of memmory

raw_data = dd.read_csv('./Resultados11.csv',
                       dtype='str',
                       assume_missing=True,
                       na_values=['', 'NA', 'nan'])

df = raw_data.compute()
df
```

```python
df = df[df['ESTU_PAIS_RESIDE'] == 'COLOMBIA']
df
```

```python
df['ESTU_DEPTO_RESIDE'].unique()
```

```python
display(df['COLE_DEPTO_UBICACION'].unique())
df['COLE_DEPTO_UBICACION'] = df['COLE_DEPTO_UBICACION'].replace({'BOGOTA': 'BOGOTÁ'})
```

```python
df['COLE_DEPTO_UBICACION'].unique()
```

```python
df = df.dropna(ignore_index=True)
```

```python
df.to_csv('./Resultados11_nonull.csv', index=False)
```

# Saber Pro

```python
raw_data = dd.read_csv('./ResultadosPro.csv',
                       dtype='str',
                       assume_missing=True,
                       na_values=['', 'NA', "nan"])

df = raw_data.compute()
df
```

```python
df = df[df['ESTU_PAIS_RESIDE'] == 'COLOMBIA']
df
```

```python
depto_col = ['BOGOTÁ', 'VALLE', 'TOLIMA', 'SANTANDER', 'SUCRE', 'ANTIOQUIA',
             'BOYACA', 'CAUCA', 'HUILA', 'CUNDINAMARCA', 'CORDOBA', 'ATLANTICO',
             'BOLIVAR', 'ARAUCA', 'NORTE SANTANDER', 'CALDAS', 'RISARALDA',
             'CESAR', 'QUINDIO', 'MAGDALENA', 'CHOCO', 'LA GUAJIRA', 'META',
             'CAQUETA', 'NARIÑO', 'PUTUMAYO', 'AMAZONAS', 'CASANARE',
             'SAN ANDRES', 'GUAVIARE', 'GUAINIA', 'VAUPES', 'VICHADA']

df = df[df['ESTU_DEPTO_RESIDE'].isin(depto_col)]
```

```python
df = df.dropna(ignore_index=True)
df
```

```python
df.to_csv('./ResultadosPro_nonull.csv')
```
